```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

df = pd.read_excel('handmade data 2.xlsx')

df = df[['order_id', 'product_name']].dropna()

total_orders = df['order_id'].nunique()
total_products = df['product_name'].nunique()

suggested_min_support = max(0.001, 2 / total_orders)

```

```python

basket = (df.groupby(['order_id', 'product_name'])['product_name']
          .count().unstack().reset_index().fillna(0)
          .set_index('order_id'))

basket = basket.map(lambda x: True if x > 0 else False)
```

```python

frequent_itemsets = apriori(basket, min_support=suggested_min_support, use_colnames=True, low_memory=True)

frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.5)

rules = rules.sort_values(['lift', 'confidence'], ascending=False)
```

```python
def get_recommendations_apriori(product_name, top_n=5, min_lift=0.5, min_confidence=0.01):
    product_frozenset = frozenset([product_name])

    relevant_rules = rules[
        (rules['antecedents'] == product_frozenset) &
        (rules['lift'] >= min_lift) &
        (rules['confidence'] >= min_confidence)
    ]

    relevant_rules = relevant_rules.sort_values(['lift', 'confidence'], ascending=False)

    recommendations = []
    for idx, row in relevant_rules.iterrows():
        rec_items = list(row['consequents'])
        for item in rec_items:
            if item != product_name:
                recommendations.append({
                    'product': item,
                    'confidence': row['confidence'],
                    'lift': row['lift'],
                    'support': row['support']
                })

    seen = set()
    unique_recs = []
    for rec in recommendations:
        if rec['product'] not in seen:
            unique_recs.append(rec)
            seen.add(rec['product'])

    return unique_recs[:top_n]
```

```python

sample_product = 'Domty - Bread - Milk Toast' #product

recommendations = get_recommendations_apriori(sample_product, top_n=10, min_lift=0.5, min_confidence=0.01)

print(f"\n'{sample_product}'By using Apriori:")
if recommendations:
    for rec in recommendations:
        print(f"- {rec['product']} (Confidence: {rec['confidence']:.3f}, Lift: {rec['lift']:.3f}, Support: {rec['support']:.4f})")
else:
    print("No Recommendation")

#Pommes Frites - Frozen Potatoes
```
